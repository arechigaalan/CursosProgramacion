```python
import pandas as pd
```

```python
 #! Importar los datos del csv, creación del df e impresión de los cinco primeros elementos. 
df = pd.read_csv('reportes_medicos.csv')
df.head(5)
```

```python
 #! Creación de la serie ´sexo/género´ y validación de nulos. 

s_genero = df['Sexo/Género']
s_genero.isnull().any()
```

```python
 #! Contar la cantidad de mujeres y hombres en el dataframe.
c_hombres = 0
c_mujeres = 0
for g in s_genero:
    if g == 'M':
        c_hombres += 1
    else:
        c_mujeres += 1

print(f'Hay {c_mujeres} mujes y {c_hombres} hombres en el dataframe.')
```

```python
 #! Creación de la serie edad.

# Cambiar el tipo de dato de la fecha.  
df['Fecha de nacimiento'] = pd.to_datetime(df['Fecha de nacimiento'])

hoy = pd.Timestamp.today()

# Cálculo de la edad y creación de la serie.
df['edad'] = ((hoy - df["Fecha de nacimiento"]).dt.days // 365)
df['edad'].head(3)
```

```python
 #! Creación de los rangos por edad

df['edad'].max() 
df['edad'].min()

df['rango edad'] = pd.cut(df['edad'], bins=range(10, 81, 10))
df['rango edad'].head(3)

```

```python
 #! Impresión de rangos y cantidades.

df["rango edad"].value_counts().sort_index()
```
